#**DIABETICS PREDICTION**

Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm # Support Vector Machine.
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

```python
diabetes = pd.read_csv("/content/diabetes.csv")
diabetes.head()
```

```python
diabetes.shape
```

```python
#Getting statistical data
diabetes.describe()
```

```python
diabetes["Outcome"].value_counts()
```

```python
diabetes.groupby("Outcome").mean()
```

```python
# Seperating data and labels
x = diabetes.drop(columns="Outcome",axis=1)
y = diabetes["Outcome"]
print(type(x))
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
standardized_data = scaler.fit_transform(x)
# standardized_data
```

```python
x1=standardized_data
y1=diabetes["Outcome"]
# Models like SVM, KNN, Logistic Regression, Neural Nets are sensitive to scale.
# Features with bigger numbers (Salary) dominate the learning.
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x1,y1,test_size=0.2,stratify=y,random_state=2)
print(xtrain.shape) # 80% for training and 20% for testing
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

Training the model

```python
classifier = svm.SVC(kernel="linear")
#SVC-> Support vector classifier
classifier.fit(xtrain,ytrain)
```

Model Evaluation

Accuracy Score

```python
# Accuracy score on the training data
xtrain_predicted = classifier.predict(xtrain)
training_data_accuracy = accuracy_score(xtrain_predicted,ytrain)
print("Accuracy score of training data : ",training_data_accuracy)
```

```python
# Accuracy score on the testing data
xtest_predicted = classifier.predict(xtest)
testing_data_accuracy = accuracy_score(xtest_predicted,ytest)
print("Accuracy score of testing data : ",testing_data_accuracy)
```

Making a Predictive System

```python
input_data= (5,166,72,19,175,25.8,0.587,51) #--Actual 1
# changing the input data to numpy array
inputnumpy = np.asarray(input_data)
# reshape the array as we are predicting for one instance
inputreshaped = inputnumpy.reshape(1,-1)

prediction = classifier.predict(inputreshaped)

if prediction[0]==0:
  print("The person is not diabetic")
else:
  print("The person is diabetic")
```

#**PROJECT-3**

#**SPAM MAIL PREDICTION**

Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

Data Collection and Pre-Processing

```python
 #Loading the data from the csv file
maildata = pd.read_csv("/content/mail_data.csv")
maildata.head()
```

```python
# Replace the null values with a null string (if in case present)
```

```python
maildata = maildata.where((pd.notnull(maildata)),'')
maildata.isnull().sum()
```

```python
maildata.shape
```

Label Encoding

```python
# Label spam mail as 0 and ham mail as 1
maildata.loc[maildata["Category"]=="spam","Category"]=0
maildata.loc[maildata["Category"]=="ham","Category"]=1
# maildata
```

```python
x = maildata["Message"]
y = maildata["Category"]
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2,random_state=3)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

Feature Extraction

```python
# transform the text data to feature vectors that can be used as input to the Logistic Regression
feature_extraction = TfidfVectorizer(min_df=1,stop_words="english",lowercase=True)
feature_extraction
# Term Frequency – Inverse Document Frequency
```

```python
x_train_features = feature_extraction.fit_transform(xtrain)
x_test_features = feature_extraction.transform(xtest)
# print(x_train_features)
# print(x_test_features)
# # uses the vocabulary learned from training data
# ignores any new words in xtest

#convert ytrain adn ytest as integers
ytrain = ytrain.astype("int")
ytest = ytest.astype("int")
```

Logistic Regression

```python
model = LogisticRegression()
```

Training Logistic Regression Model With the Training Data

```python
model.fit(x_train_features,ytrain)
```

Evaluating the training model

```python
# Prediction on training data
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(ytrain,prediction_on_training_data)
print("Accuracy on training data : ",accuracy_on_training_data)
```

```python
#Predicting on testing data
prediction_on_testing_data = model.predict(x_test_features)
accuracy_on_testing_data = accuracy_score(ytest,prediction_on_testing_data)
print("Accuracy on testing data : ",accuracy_on_testing_data)
```

Building a predective system

```python
input_mail = ["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry "]
# Convert text to feature_vectors

inputfeatures = feature_extraction.transform(input_mail)
# print(inputfeatures)

# print(prediction)
prediction = model.predict(inputfeatures)
if prediction[0]==0:
  print("Spam Mail")
else:
  print("Ham Mail")
```

```python
input_mail = ["Haha yeah I see that now, be there in a sec"]
# Convert text to feature_vectors

inputfeatures = feature_extraction.transform(input_mail)
# print(inputfeatures)

# print(prediction)
prediction = model.predict(inputfeatures)
if prediction[0]==0:
  print("Spam Mail")
else:
  print("Ham Mail")
```
